```python
from reportlab.lib.pagesizes import letter
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer
from reportlab.lib.styles import getSampleStyleSheet
from reportlab.pdfgen import canvas

def create_pdf():
    pdf_file = "output.pdf"
    document = SimpleDocTemplate(pdf_file, pagesize=letter)

    # Create a list to hold flowable elements
    story = []

    # Set up styles
    styles = getSampleStyleSheet()
    normal_style = styles["Normal"]

    # Example content
    main_heading_text = "Collge Name"
    sub_heading_text = "Course Name"
    section_title_text = "Section 1"
    time_text = "Time: 1 hour"
    marks_text = "Marks: 20"
    question_text = "What is the capital of France?"

    # Main heading
    main_heading = Paragraph(main_heading_text, styles["Heading1"])
    story.append(main_heading)

    # Sub heading
    sub_heading = Paragraph(sub_heading_text, styles["Heading2"])
    story.append(sub_heading)

    # Section 1
    story.append(Spacer(1, 20))  # Add space
    story.append(Paragraph(section_title_text, styles["Heading3"]))
    
    # Draw time and marks
    c = canvas.Canvas(pdf_file, pagesize=letter)
    c.setFont("Helvetica", 12)
    time_x = 50
    marks_x = letter[0] - c.stringWidth(marks_text, "Helvetica", 12) - 50
    y_position = letter[1] - 50
    c.drawString(time_x, y_position, time_text)
    c.drawString(marks_x, y_position, marks_text)
    c.save()

    # Question
    story.append(Paragraph(question_text, normal_style))

    # Build the PDF document
    document.build(story)

create_pdf()
```

Prompt: How to center the heading using reportlab in python in an A4 sheet

```python
from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import A4
from reportlab.lib import colors

def center_text(canvas, text, font_size=12):
    width, height = A4
    text_object = canvas.beginText((width - canvas.stringWidth(text, 'Helvetica', font_size)) / 2, height - 50)
    text_object.setFont('Helvetica', font_size)
    text_object.textLine(text)
    canvas.drawText(text_object)

def generate_pdf(file_path, heading):
    c = canvas.Canvas(file_path, pagesize=A4)
    
    # Set font and size for the heading
    c.setFont("Helvetica", 16)
    
    # Call the center_text function to center the heading
    center_text(c, heading, font_size=16)

    # Save the canvas to the file
    c.save()

# Replace 'output.pdf' with your desired file name and 'Your Heading' with your actual heading
generate_pdf('output.pdf', 'Your Heading')
```

```python
from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import A4
from reportlab.lib import colors

def center_text(canvas, text, font_size=12):
    width, height = A4
    text_object = canvas.beginText((width - canvas.stringWidth(text, 'Helvetica', font_size)) / 2, height - 50)
    text_object.setFont('Helvetica', font_size)
    text_object.textLine(text)
    canvas.drawText(text_object)

def generate_question_paper(file_path, heading1, heading2, heading3, time, marks):
    c = canvas.Canvas(file_path, pagesize=A4)
    
    # Set font and size for the headings
    c.setFont("Helvetica", 16)
    
    # Center-aligned headings
    center_text(c, heading1, font_size=16)
    center_text(c, heading2, font_size=16)
    center_text(c, heading3, font_size=16)
    
    # Left-aligned time
    c.setFont("Helvetica", 12)
    c.drawString(50, A4[1] - 100, f"Time: {time}")

    # Right-aligned marks
    c.drawRightString(A4[0] - 50, A4[1] - 100, f"Marks: {marks}")

    # Save the canvas to the file
    c.save()

# Replace 'question_paper.pdf' with your desired file name and provide the necessary information
generate_question_paper('output.pdf', 'Heading 1', 'Heading 2', 'Heading 3', '2 hours', '100')
```
