```python
from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import A4
from reportlab.lib import colors

def center_text(canvas, text, font_size=12):
    width, height = A4
    text_object = canvas.beginText((width - canvas.stringWidth(text, 'Helvetica', font_size)) / 2, height - 50)
    text_object.setFont('Helvetica', font_size)
    text_object.textLine(text)
    canvas.drawText(text_object)

def generate_pdf(file_path, heading):
    c = canvas.Canvas(file_path, pagesize=A4)
    
    # Set font and size for the heading
    c.setFont("Helvetica", 16)
    
    # Call the center_text function to center the heading
    center_text(c, heading, font_size=16)

    # Save the canvas to the file
    c.save()

# Replace 'output.pdf' with your desired file name and 'Your Heading' with your actual heading
generate_pdf('output.pdf', 'Your Heading')
```
